## Structure Builder

The [Structure Builder](https://lab.cp2k.com/docs/structure-builder) is an app for creating and manipulating atomic structures. We will start by importing an H2O molecule from the database. To make things more intersting, we're then going to distort its geometry a bit. And finally we'll move on the CP2K-Wizard.

### Import a structure

1. Click on **Import molecule** to open the import dialog.
2. In dialog click on the **SEARCH** tab.
3. Search for *"H2O"* and click **SELECT**
4. In the following dialog fill in *"distorted_h2o"* as new name and click **RENAME**.

### Change a bond length

1. Select the **Adjust bond length** tool from the toolbar.
2. Click on one of the Hydrogen atoms and then on the Oxygen atoms.
3. Increase the bond length to around 2 Angstr√∂m.
4. Click on **APPROVE**.

### Change a bonds angle
1. Select the **Adjust bond angle** tool from the toolbar.
2. Click on first Hyrdogen, then on the Oxygen, and finally on the other Hydrogen.
3. Increase the bonds angle to around 170 degrees.
4. Click on **APPROVE**.
  
### Go to the CP2K Wizard
1. Once you're satisfied with the molecule's geometry, click on **OPEN WIZARD** at the bottom.






## CP2K Wizard

The [CP2K Wizard](https://lab.cp2k.com/docs/cp2k-wizard) is an app to quickly setup common simulations. We'll run a geometry optimization to compute the ground state geometry of the H2O molecule.

1. Open the **Force Eval** section and increase the **Max SCF Iterations** to **100**. Return the overview by clicking **BACK** in the top left corner.
2. Open the **Run Type** section and select **Geometry Optimization**.
3. Open the **Finish** section and select **Machine Size** the be **Normal**.
4. Back at the main view, click on **RUN INPUT**.
5. In the appearing popup dialog, click on **RENAME** to confirm the suggested filename.

This will launch a compute job and take you to the corresponding output viewer.


## Output Browser

The [Output Browser](https://lab.cp2k.com/docs/output-browser) shows an individual calculation including its inputs, outputs, and key results.


1. Wait for the job to start (might take 2 minutes).
2. Watch the energy plots of the outer geometry optimization and the inner SCF optimization.
3. Wait for the calculation to finish (should take around 1 minute).
4. Click on **Run1-pos-1.xyz** in right sidebar to open the trajectory viewer.


## Tracjectory Viewer

The [Trajectory Viewer](https://lab.cp2k.com/docs/trajectory-viewer) is an app for viewing atomic motion.



1. View the trajectory a few times from different angles.
2. Note how the optimization typically overshoots a bit before settling in. This corresponds to spikes in the total energy plot that you might have noticed earlier.
3. Goto the last frame, which contains the ground state structure.
4. Open the three-dots menu on the right and select **Export current frame**, which open the structure builder again.
5. In the structure builder click directly on **OPEN WIZARD**.

## Compute Properties

This time we'll use the Wizard to setup a single point calculation to compute several ground state properties.

1. Open the **Run Type** section and select **Write dipole moment**, **Write orbitals as .cube files**, and **Write Hartree potential as .cube**.
2. Open the **Finish** section and select **Machine Size** the be **Normal**.
3. Back at the main view, click on **RUN INPUT**.
4. In the appearing popup dialog, chose an appropriate name, e.g. *"H2O-properties"*, and click **RENAME** to confirm.
5. Wait for the calculation to finish.
6. Note down the **Dipol Moment**, which you'll find in the right sidebar. Clicking on it will highlight the corresponding output lines.

## View Hartee Potential

The [CUBE Viewer](https://lab.cp2k.com/docs/cube-viewer) is an app to view iso-surfaces of volumetric datasets.

1. Click on **Run1-v_hartree-1_0.cube** to view the electric potential.
2. Note the two lone electron pairs that give rise to the strong dipole moment of the H2O molecule.
3. Try different isolevels.

Repeat the above steps for the following molecules:

2. H2 (hydrogen molecule)
3. CH4 (methane)
4. O2 (triplet oxygen)
5. hexatriene (C6H8)

For O2 you also need to set the **Spin Multiplicity** to **Triplet** in the **Force Eval** section.
